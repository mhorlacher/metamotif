```python
import metamotif
```

### Load example data of feature importances calculated on a regression task to predict QKI binding via sequence

```python
sequence_importances = metamotif.example_data.QKI
```

### Inspect the center of raw sequence importance scores, where we expect the binding motifs. 

```python
for si in sequence_importances[:3]:
    metamotif.plot_motif(si[150:250], sigma=['A', 'C', 'G', 'U'])
```

### We can clearly see the canonical QKI motif (AC)UAA!

### Let's extract meta-motifs of size k=5 from the sequences

```python
meta_motifs = metamotif.extract_meta_motifs(sequence_importances, k=5)
meta_motifs.shape
```

### We can see that the algorithm extracted 4 meta motifs. Let's visualize them!

```python
for m in meta_motifs:
    metamotif.plot_motif(m, sigma=['A', 'C', 'G', 'U'])
```

### As we can see, the algorithm successfully recovers the canonical motif of QKI, however, we also see that a characteristic U-rich bias-motif was extract, which is due to biases in the underlying data (eCLIP).
